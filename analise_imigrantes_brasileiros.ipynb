# **Aula 1 - Conhecendo a biblioteca Matplotlib**


##**Conhecendo os dados**

### Importando os dados

```python
import pandas as pd
```

```python
df = pd.read_csv('imigrantes_canada.csv')
df.info()
```

### Realizando alguns tratamentos

```python
df.set_index('País', inplace=True)
anos = list(map(str, range(1980, 2014)))
anos
```

### Extraindo a série de dados para o Brasil

```python
brasil = df.loc['Brasil', anos]
brasil_dict = {'ano': brasil.index.tolist(), 'imigrantes': brasil.values.tolist()}

dados_brasil = pd.DataFrame(brasil_dict)
dados_brasil
```

```python
import matplotlib.pyplot as plt
```

## Evolução de imigrantes brasileiros para o Canadá

```python
fig, ax = plt.subplots()
ax.plot(dados_brasil['ano'], dados_brasil['imigrantes'], lw=3)
ax.xaxis.set_major_locator(plt.MultipleLocator(5))

ax.set_title('Imigrantes do Brasil para o Canadá (1980-2013)', fontsize=14)
ax.set_xlabel('Ano', fontsize=12)
ax.set_ylabel('Número de Imigrantes', fontsize=12)

ax.yaxis.set_tick_params(labelsize=10)
ax.xaxis.set_tick_params(labelsize=10)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.show()
```

```python
# gráfico comparativo dos países da américa do sul e o canadá

paises_sul_america = ['Brasil', 'Argentina', 'Colômbia', 'Chile', 'Peru', 'Venezuela', 'Uruguai', 'Paraguai', 'Equador', 'Canadá']
df_sul_america = df.loc[paises_sul_america, anos].transpose()
df_sul_america.head()

fig, ax = plt.subplots(figsize=(10, 6))
for pais in paises_sul_america:
    ax.plot(df_sul_america.index, df_sul_america[pais], lw=2, label=pais)
ax.xaxis.set_major_locator(plt.MultipleLocator(5))
ax.set_title('Imigrantes para o Canadá de Países da América do Sul (1980-2013)', fontsize=14)
ax.set_xlabel('Ano', fontsize=12)
ax.set_ylabel('Número de Imigrantes', fontsize=12)
ax.legend()
plt.show()
```
